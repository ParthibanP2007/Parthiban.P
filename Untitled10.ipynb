```python
# 1. Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_digits
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
# 2. Load Dataset (Digits dataset as example - 8x8 images of numbers 0-9)
digits = load_digits()
X = digits.data  # Flattened images (1797 x 64)
y = digits.target
images = digits.images  # Original 8x8 images

print("Dataset Shape:", X.shape)
print("Number of Images:", len(images))
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Explained variance ratio
explained_var = np.cumsum(pca.explained_variance_ratio_)

# Plot variance explained vs number of components (Elbow-like method for PCA)
plt.figure(figsize=(8,5))
plt.plot(range(1, len(explained_var)+1), explained_var, marker='o')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Choosing Optimal Number of Components")
plt.grid(True)
plt.show()
```

```python
# 5. Choose optimal number of components (e.g., 30 for ~95% variance)
pca_opt = PCA(n_components=30)
X_pca_opt = pca_opt.fit_transform(X_scaled)

print("Original Shape:", X_scaled.shape)
print("Reduced Shape:", X_pca_opt.shape)
print("Variance Retained:", np.sum(pca_opt.explained_variance_ratio_)*100, "%")
```

```python
# 6. Reconstruct Images from Reduced Dimensions
X_reconstructed = pca_opt.inverse_transform(X_pca_opt)

# Compute reconstruction error (MSE)
error = mean_squared_error(X_scaled, X_reconstructed)
print("Reconstruction Error (MSE):", error)
```

```python
n = 10  # number of samples
plt.figure(figsize=(20,4))
for i in range(n):
    # Original
    plt.subplot(2, n, i+1)
    plt.imshow(images[i], cmap="gray")
    plt.title("Original")
    plt.axis("off")

    # Reconstructed
    plt.subplot(2, n, i+1+n)
    plt.imshow(X_reconstructed[i].reshape(8,8), cmap="gray")
    plt.title("Reconstructed")
    plt.axis("off")
plt.show()
```

```python
# 8. Visualize PCA Components (Eigenvectors)
plt.figure(figsize=(10,5))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(pca_opt.components_[i].reshape(8,8), cmap="gray")
    plt.title(f"PC {i+1}")
    plt.axis("off")
plt.suptitle("Top 10 Principal Components (Eigenvectors)")
plt.show()
```

```python
# 9. Answering Questions with Code + Metrics
print("✅ Optimal Components (by 95% variance):", 30)
print("✅ Variance Retained:", round(np.sum(pca_opt.explained_variance_ratio_)*100,2), "%")
print("✅ Reconstruction Error:", round(error,4))
print("✅ PCA reduces dimensions from 64 → 30, improving speed in ML models.")
```
