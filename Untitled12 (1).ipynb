```python
# 1. Import Libraries
# ==============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
# 2. Create Dataset & Save to CSV
# ==============================
data = {
    "Alcohol": [13.2, 12.4, 14.1, 13.8, 12.1, 11.9, 13.5, 14.3, 12.9, 13.1],
    "Malic_Acid": [1.8, 2.1, 1.6, 2.4, 1.9, 2.5, 1.7, 1.5, 2.3, 2.0],
    "Ash": [2.4, 2.2, 2.5, 2.1, 2.3, 2.0, 2.6, 2.4, 2.2, 2.3],
    "Alcalinity": [19, 18, 20, 16, 17, 15, 19, 21, 18, 20],
    "Magnesium": [100, 95, 110, 102, 98, 92, 105, 108, 99, 103],
    "Phenols": [2.8, 2.5, 3.1, 2.9, 2.7, 2.3, 3.0, 2.6, 2.4, 2.9],
    "Flavanoids": [2.4, 2.1, 2.8, 2.6, 2.3, 1.9, 2.7, 2.2, 2.0, 2.5],
    "WineType": [0, 1, 2, 0, 1, 2, 0, 2, 1, 0]  # 0,1,2 â†’ three wine classes
}

df = pd.DataFrame(data)
df.to_csv("wine_dataset.csv", index=False)
print("Dataset saved as wine_dataset.csv")
```

```python
# 3. Load Dataset
# ==============================
df = pd.read_csv("wine_dataset.csv")
print(df.head())

```

```python
# 4. Feature Matrix and Labels
# ==============================
X = df.drop("WineType", axis=1)
y = df["WineType"]
```

```python
# 5. Feature Scaling (important for SVM)
# ==============================
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# 6. Train-Test Split
# ==============================
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42
)
```

```python
# 7. Decision Tree Model
# ==============================
dt = DecisionTreeClassifier(max_depth=3, random_state=42)
dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)

print("\nDecision Tree Performance:")
print(classification_report(y_test, y_pred_dt))

# Feature Importance
feature_importance = pd.Series(dt.feature_importances_, index=X.columns)
print("\nFeature Importance (Decision Tree):")
print(feature_importance.sort_values(ascending=False))

```

```python
# 8. Support Vector Machine Model
# ==============================
svm = SVC(kernel="rbf", C=1, gamma="scale", probability=True)
svm.fit(X_train, y_train)

y_pred_svm = svm.predict(X_test)

print("\nSVM Performance:")
print(classification_report(y_test, y_pred_svm))
```

```python
# 9. Confusion Matrices
# ==============================
cm_dt = confusion_matrix(y_test, y_pred_dt)
cm_svm = confusion_matrix(y_test, y_pred_svm)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))
sns.heatmap(cm_dt, annot=True, fmt="d", cmap="Blues", ax=ax[0])
ax[0].set_title("Decision Tree Confusion Matrix")
sns.heatmap(cm_svm, annot=True, fmt="d", cmap="Greens", ax=ax[1])
ax[1].set_title("SVM Confusion Matrix")
plt.show()
```

```python
# 10. Cross-Validation (for accuracy)
# ==============================
cv_dt = cross_val_score(dt, X_scaled, y, cv=3).mean()
cv_svm = cross_val_score(svm, X_scaled, y, cv=3).mean()

print(f"\nCross-Validation Accuracy - Decision Tree: {cv_dt:.2f}")
print(f"Cross-Validation Accuracy - SVM: {cv_svm:.2f}")
```

```python
# 11. Hyperparameter Tuning (GridSearch for SVM)
# ==============================
param_grid = {
    "C": [0.1, 1, 10],
    "gamma": ["scale", 0.1, 0.01],
    "kernel": ["linear", "rbf", "poly"]
}
grid = GridSearchCV(SVC(), param_grid, refit=True, cv=3)
grid.fit(X_train, y_train)

print("\nBest SVM Parameters:", grid.best_params_)
```

```python
# 12. Visualize Decision Tree
# ==============================
plt.figure(figsize=(12, 6))
plot_tree(dt, feature_names=X.columns, class_names=["Class0", "Class1", "Class2"], filled=True)
plt.show()
```

```python
# 13. Visualize Decision Boundaries (only with 2 features)
# ==============================
def plot_decision_boundary(model, X, y, title):
    X = X[:, :2]  # take only first 2 features
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                         np.linspace(y_min, y_max, 200))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
    plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.coolwarm, edgecolors="k")
    plt.title(title)
    plt.show()

# Train SVM on only 2 features for visualization
svm_vis = SVC(kernel="rbf", C=1, gamma="scale")
svm_vis.fit(X_train[:, :2], y_train)
plot_decision_boundary(svm_vis, X_train, y_train, "SVM Decision Boundary (2 features)")
```
